```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.isna().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['JobSatisfaction'].fillna(df['JobSatisfaction'].median(), inplace=True)
```

```python
df['MonthlyIncome'].fillna(df['MonthlyIncome'].median(), inplace=True)
```

```python
df['OverTime'].fillna(df['OverTime'].mode()[0], inplace=True)
```

```python
df['YearsAtCompany'].fillna(df['YearsAtCompany'].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['MonthlyIncome'] < lower) | (df['MonthlyIncome'] > upper)]

print(outliers)
```
